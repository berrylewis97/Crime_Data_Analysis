```python
#see corelation between age and area
label = ['0-15','16-25','26-35','36-45','46-55','56-65','66-75','76-85','86-95']
bins = [0, 16, 26, 36, 46, 56, 66, 76, 86, 96]
data['age_range'] = pd.cut(data['Vict_Age'],bins = bins, labels=label )

contingency = pd.crosstab(data['Area_Name'], data['age_range'])

plt.figure(figsize=(12, 8))
sns.heatmap(contingency, annot=True, cmap="YlGnBu", fmt='g')

plt.title("Contingency table between Area_Name and age_range")
plt.show()
```
